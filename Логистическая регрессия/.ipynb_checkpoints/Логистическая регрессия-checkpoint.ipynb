```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LinearRegression
```

```python
X = np.random.sample((200, 5))
y = np.random.randint(0,2, (200,1))
X2 = np.random.sample((1, 5))
```

```python
X
```

```python
ones = np.ones((X.shape[0], 1))
```

```python
X1 = np.hstack((ones, X))
```

```python
XTX = X1.T.dot(X1)
```

```python
XTX_inv = np.linalg.inv(XTX)
```

```python
b = XTX_inv.dot(X1.T).dot(y)
```

```python
b
```

```python
ones = np.ones((1, 1))
```

```python
X3 = np.hstack(((ones, X2)))
```

```python
z = X3.dot(b)
```

```python
z1 = 1/(1+np.exp(-z))
```

```python
z1
```

```python
z
```

```python
lr = LogisticRegression()
```

```python
lr.fit(X,y)
```

```python
lr.predict_proba(X2)
```
